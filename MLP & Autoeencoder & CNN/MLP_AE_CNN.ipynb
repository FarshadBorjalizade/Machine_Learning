## MLP

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# scale data
t = StandardScaler()
t.fit(X_train)
X_train = t.transform(X_train)
X_test = t.transform(X_test)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(20,),
                    solver='adam',
                    activation='relu',
                    batch_size=32,
                    learning_rate_init = 0.0001, 
                    max_iter=400, verbose=True, 
                    early_stopping=True, random_state=42)
```

```python
mlp.fit(X_train,y_train)
```

```python
predictions = mlp.predict(X_test)
```

```python
print('confusion matrix: \n',confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
X_train = train_data.iloc[:, 0:54]
y_train = train_data['label'].astype(np.int32)
print(X_train.shape, y_train.shape)
```

```python
X_test = test_data.iloc[:, 0:54]
y_test = test_data['label'].astype(np.int32)
print(X_test.shape, y_test.shape)
```

```python
X_train.min(), X_train.max()
```

```python
X_test.min(), X_test.max()
```

```python
# scale data
t = MinMaxScaler()
t.fit(X_train)
X_train = t.transform(X_train)
X_test = t.transform(X_test)
```

```python
X_train.min(), X_train.max()
```

```python
X_test.min(), X_test.max()
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# define encoder
n_inputs = X_train.shape[1]
visible = Input(shape=(n_inputs,))

# encoder level 1
e = Dense(n_inputs*2)(visible)
e = BatchNormalization()(e)
e = LeakyReLU()(e)

# encoder level 2
e = Dense(n_inputs)(e)
e = BatchNormalization()(e)
e = LeakyReLU()(e)

# bottleneck
n_bottleneck = n_inputs
bottleneck = Dense(n_bottleneck)(e)

# define decoder, level 1
d = Dense(n_inputs)(bottleneck)
d = BatchNormalization()(d)
d = LeakyReLU()(d)

# decoder level 2
d = Dense(n_inputs*2)(d)
d = BatchNormalization()(d)
d = LeakyReLU()(d)

# output layer
output = Dense(n_inputs, activation='relu')(d)

# define autoencoder model
autoencoder_model = Model(inputs=visible, outputs=output)
autoencoder_model.summary()
```

```python
# compile autoencoder model
autoencoder_model.compile(optimizer='adam', loss='mse')

# fit the autoencoder model to reconstruct input
history = autoencoder_model.fit(X_train, X_train, epochs=20, batch_size=16, verbose=2,
                    callbacks=[early_stop], validation_data=(X_test, X_test))
```

```python
model_history = pd.DataFrame(autoencoder_model.history.history)
```

```python
# summarize history for loss
model_history[['loss', 'val_loss']].plot()
```

```python
autoencoder_model.evaluate(X_test, X_test)
```

## CNN Model

```python
X_train = train_data.iloc[:, 0:54]
y_train = train_data['label'].astype(np.float32)
print(X_train.shape, y_train.shape)
```

```python
X_test = test_data.iloc[:, 0:54]
y_test = test_data['label'].astype(np.float32)
print(X_test.shape, y_test.shape, y_test.dtype)
```

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Create cnn model
n_input = Input(shape=(54,))

first_layer = Dense(units=64, activation='relu')(n_input)
first_layer = BatchNormalization()(first_layer)
first_layer = LeakyReLU()(first_layer)

second_layer = Dense(units=32, activation='relu')(first_layer)
second_layer = BatchNormalization()(second_layer)
second_layer = LeakyReLU()(second_layer)

third_layer = Dense(units=20, activation='relu')(second_layer)
third_layer = BatchNormalization()(third_layer)
third_layer = LeakyReLU()(third_layer)

forth_layer = Dense(units=10, activation='relu')(third_layer)
forth_layer = BatchNormalization()(forth_layer)
forth_layer = LeakyReLU()(forth_layer)

fifth_layer = Dense(units=5, activation='relu')(forth_layer)
fifth_layer = BatchNormalization()(fifth_layer)
fifth_layer = LeakyReLU()(fifth_layer)

output_layer = Dense(units=2, activation='sigmoid')(fifth_layer)

simple_cnn_model = Model(inputs=n_input, outputs=output_layer)
simple_cnn_model.summary()
```

```python
# compile the model
simple_cnn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# fit model on the dataset
simple_cnn_model.fit(X_train, y_train, epochs=40, batch_size=36,
                    validation_data=(X_test, y_test), verbose=2,
                    callbacks=[early_stop])
```

```python
model_history = pd.DataFrame(simple_cnn_model.history.history)
```

```python
# summarize history for loss
model_history[['loss', 'val_loss']].plot()
```

```python
# summarize history for accuracy
model_history[['accuracy', 'val_accuracy']].plot()
```

```python
simple_cnn_model.evaluate(X_test, y_test)
```
